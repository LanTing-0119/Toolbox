```python
import os
from pathlib import Path

def create_file_dict(root_dir, suffix):
    """
    创建一个字典，键是文件名(不带后缀)，值是文件的绝对路径
    
    参数:
        root_dir (str): 要遍历的根目录路径
        
    返回:
        dict: {文件名: 绝对路径} 的字典
    """
    file_dict = {}
    
    for root, _, files in os.walk(root_dir):
        for file in files:
            if file.startswith('.'):
                continue
                
            # 获取文件绝对路径
            file_path = os.path.abspath(os.path.join(root, file))
            # print(file_path)
            
            # 获取不带后缀的文件名: 02体格检查-胸部检查-08.肺部叩诊的方法和顺序
            file_name = Path(file).stem


            
            if suffix == True:
                # 添加到字典
                file_dict[file_name] = file_path
            if suffix == False:
                # 添加到字典
                if file_path[-4:]=='.mp4':
                    file_dict[file_name] = file_path.replace('.mp4','')
    
    return file_dict

srts_dir = '/Volumes/岚霆_临床/[]srt_cn_绍绍'
dict_srts = create_file_dict(srts_dir, suffix = True)

mp4_dir = '/Volumes/岚霆_临床/视频_学习/校内医学参考/完整教学视频/绍绍医考_理论'
dict_mp4s = create_file_dict(mp4_dir, suffix = False)
```

```python
import shutil
import os

def copy_common_files(dict_a, dict_b):
    # 获取两个字典的共有键
    common_keys = set(dict_a.keys()) & set(dict_b.keys())
    i=0
    
    for key in common_keys:
        src_path = dict_a[key]
        dest_path = dict_b[key] + '.srt'
        
        # 检查源文件是否存在
        if not os.path.isfile(src_path):
            print(f"警告：源文件 '{src_path}' 不存在，跳过。")
            continue
        
        # 获取目标目录并创建（如果需要）
        dest_dir = os.path.dirname(dest_path)
        if dest_dir:  # 防止空目录（例如当前目录）
            os.makedirs(dest_dir, exist_ok=True)
        
        # 复制文件
        try:
            shutil.copy2(src_path, dest_path)
            print(f"已复制：'{src_path}' -> '{dest_path}'")
            i+=1
        except PermissionError:
            print(f"错误：没有权限写入 '{dest_path}'")
        except Exception as e:
            print(f"复制文件时发生错误：{e}")

    print(i)

copy_common_files(dict_srts, dict_mp4s)
```
